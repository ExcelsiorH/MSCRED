```python
import pandas as pd

file_path = r"data\origin_data\均一性测试-3\20250712\24r 10min\20250712_24r10min.xlsx"
df1 = pd.read_excel(file_path)
df1.head()
```

```python
len(df1)
```

```python
file_path = r"data\origin_data\均一性测试-3\20250712\24r 15min\20250712_24r15min.xlsx"
df2 = pd.read_excel(file_path)
df2.head()
```

```python
# 将df1和df2合并成一个DataFrame
df_combined = pd.concat([df1, df2], ignore_index=True)
df_combined.head()
```

```python
len(df_combined)
```

```python
#将df_combined去掉timestamp列
if 'timestamp' in df_combined.columns:
    df_combined = df_combined.drop(columns=['timestamp'])
```

```python
df_combined.head()
```

```python
#将df_combined转置
df_combined_transposed = df_combined.T
df_combined_transposed.head()
#将df_combined_transposed保存为csv文件
output_file_path = r"data\origin_data\uniformity_test3.csv"
df_combined_transposed.to_csv(output_file_path, index=False, header=False)
```
